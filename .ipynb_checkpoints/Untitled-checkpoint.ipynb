```python
import pandas as pd
import glob
import os
```

```python
#посмотрели  сколько у нас csv файлов 
files = glob.glob("*.csv")
files
```

```python
# комбинируем наши csv файлы в один dataframe  по имени combined 
combined = pd.DataFrame() 
for file in files:
    data = pd.read_csv(file,sep = ";")
    data["filname"] = file
    combined = pd.concat([combined,data],axis = 0, ignore_index = True )
```

```python
# смотрим наш уже комбинированный  dataframe
combined 
# видим что у нас 149908 строк и с файлнеймами 13 колонок 
```

```python
# спомошью drop и duplicates очищаем наш dataframe от дуьликотов 
# оставляем наш data frame без дубликатов для дольнейшей работы 
combined = combined.drop_duplicates()
combined
# видим что у нас остались 134061 строк и соответсвенно 13 колонок 
```
