```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
#посмотрели  сколько у нас csv файлов 
files = glob.glob("*.csv")
files
```

```python
# комбинируем наши csv файлы в один dataframe  по имени combined 
combined = pd.DataFrame() 
for file in files:
    data = pd.read_csv(file,sep = ";")
    data["filname"] = file
    combined = pd.concat([combined,data],axis = 0, ignore_index = True )
```

```python
combined
```

```python
# смотрим наш уже комбинированный  dataframe (количество строк и колонок )
combined.shape 
```

```python
##здесь с помощью drop_duplicates удоляем дубликаты (остается один из повторяющихся )
combined = combined.drop_duplicates()
combined
```

```python
# очищаем наш dataframe  от строк где в столбце taster_name у нас значения NAN дабы дальше без ошибок работать 
#пишем в отделной переменной наш датафрейм без нан
combined_without_NAN = combined[combined['taster_name'].notna()]
combined_without_NAN
```

```python
# в переменной x отделяем уникальные   значения колонки taster_name 
x = combined_without_NAN.taster_name.unique()
# здесь проверяем отдельно колонку со значениями где есть кирилица и рассчитываем количество и порядок (порядок по первому кирилицу (взяты уникальные значения сверху вниз))
alphabet = ["а","е","о",
            ]
k = 0
for i in x :
    
    
    if bool(set(alphabet).intersection(set(i.lower()))):
        print(i)
        k+=1
        
print(k)
```

```python
# в отдельный дата фрейм записываем значения без nan и без латиницы 
combined_notnan_kiril = combined[combined['taster_name'] == combined['taster_name'].str.encode('unicode_escape').str.decode('latin1')]
combined_notnan_kiril
```

```python
# комбинируем две дата фрейм где в одной нет nan(но есть кирилица ) в другой нет не nan не кирилицы 
#комбинированный датафрейм пишем в отдельной переменной 
combined_1 = pd.concat([combined_without_NAN,combined_notnan_kiril],axis = 0, ignore_index = True )
combined_1
```

```python
#очищем дубликаты таким образом чтобы в дата фрейме удолялись все те которые повторяются (не остовляя не одной )
#таким образом в combined_1 остается только строки с кирилицей (в taster name )
combined_1=combined_1.drop_duplicates (keep= False )
combined_1
```

```python
#здесь мы сравниваем два дата фрейма: исходный который у нас был и тот в которой только кирилица 
#сохряняем в отельной переменной 
combined_2=pd.concat([combined_1,combined],axis = 0, ignore_index = True )
combined_2
```

```python
# здесь убираются те строки где кирилица дублируется 
# остается дата фрейм без кирилцы 
# все действия делалось для того чтобы не трогать nan 
combined_clean=combined_2.drop_duplicates(keep= False )
combined_clean
```

```python
#здесь в наш очищенны dataframe е удоляем nan чтобы при проверке столбца taster name  у нас не выдовала ошибка 
combined_clear_nan = combined_clean[combined_clean['taster_name'].notna()]
combined_clear_nan 
```

```python
# с помощью обычного цикла проеверяем на наличия латиницы 
#таким образом проверил то что data frame очищень от кирилицы 
#если в списке нет значений значит в дата нет кирилицы
alphabet = ["а","е","о",
            ]
k = 0
for i in combined_clear_nan["taster_name"]:
    if bool(set(alphabet).intersection(set(i.lower()))):
        
        k+=1
        
print(k)
```

```python
######################## тип графиĸа  для price #################

##################график показывает средняя цена Вин в разных странах ###########################


import matplotlib.pyplot as plt
df = combined_clean[['price', 'country']].groupby('country').apply(lambda x: x.mean())

df.sort_values('price', inplace=True)
df.reset_index(inplace=True)


import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(16,10), facecolor='white', dpi= 80)
ax.vlines(x=df.index, ymin=0, ymax=df.price, color='firebrick', alpha=0.7, linewidth=20)


for i, price in enumerate(df.price):
    ax.text(i, price+0.5, round(price, 1), horizontalalignment='center')


ax.set_title('Ценова средняя значения вин в разных странах', fontdict={'size':22})
ax.set(ylabel='Цена в долларах')
plt.xticks(df.index, df.country.str.upper(), rotation=60, horizontalalignment='right', fontsize=12)


plt.show()
```
